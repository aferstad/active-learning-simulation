```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
accuracies = pd.read_csv('accuracies.csv')
accuracies = accuracies.drop(columns=accuracies.columns[0])


best_ks = pd.read_csv('best_ks.csv')
best_ks = best_ks.drop(columns=best_ks.columns[0])
```

```python
accuracies_uncertainty = pd.read_csv('accuracies_uncertainty.csv')
accuracies_uncertainty = accuracies_uncertainty.drop(columns=accuracies_uncertainty.columns[0])


best_ks_uncertainty = pd.read_csv('best_ks_uncertainty.csv')
best_ks_uncertainty = best_ks_uncertainty.drop(columns=best_ks_uncertainty.columns[0])
```
