## Bayesian Approach

```python
from input.heart_import import get_heart_data
from sklearn.linear_model import LogisticRegression

import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

import pymc3 as pm

import matplotlib.pyplot as plt
import pandas as pd
```

```python
def split_test_train(df, n_components = 2, random_state = 0, do_pca = True):
    test = df.sample(frac=0.2, random_state=random_state)
    train = df.drop(test.index)

    X_train = train.drop(train.columns[0], axis=1)
    y_train = train.iloc[:, 0]

    X_test = test.drop(test.columns[0], axis=1)
    y_test = test.iloc[:, 0]   

    scaler = StandardScaler()
    pca = PCA(n_components=n_components)

    scaler.fit(X_train)
    X_train = scaler.transform(X_train)
    X_test = scaler.transform(X_test)
    
    if do_pca:
        pca.fit(X_train)
        print(sum(pca.explained_variance_ratio_))

        X_train = pca.transform(X_train)
        X_test = pca.transform(X_test)

    # Add columns of ones to X matrices to simplify fitting of intercept:
    ones = np.array([1] * X_train.shape[0])
    X_train = np.column_stack((ones, X_train))

    ones = np.array([1] * X_test.shape[0])
    X_test = np.column_stack((ones, X_test))
    
    y_test = y_test.reset_index(drop=True)
    y_train = y_train.reset_index(drop=True)

    return X_test, y_test, X_train, y_train
```

```python
def create_subsets(X, y):
    X = pd.DataFrame(X)
    y = pd.DataFrame(y)
    data = pd.DataFrame(X)
    data.insert(0, 'y', y)
    
    subsets = []

    for i in range(6):
        subset = data.sample(n=40, random_state = i*2)
        subsets.append(subset)
        data.drop(subset.index)

    return subsets
```

```python
df = get_heart_data()
df.shape
```

```python
X_test, y_test, X_train, y_train = split_test_train(df, do_pca = False)
subsets = create_subsets(X_train, y_train)
```

```python
import matplotlib as mpl
from scipy import stats


print('Posterior distributions after ' + str(len(traces)) + ' iterations.')

cmap = mpl.cm.winter


components = traces[0]['betas'].shape[1]


for component in range(components):
    
    plt.figure(figsize=(8, 2))
    
    for update_i, trace in enumerate(traces):
        samples = trace['betas'][:,component]
        smin, smax = np.min(samples), np.max(samples)
        x = np.linspace(smin, smax, 100)
        y = stats.gaussian_kde(samples)(x)
        plt.plot(x, y, color=cmap(1 - update_i / len(traces)))
    #plt.axvline({'alpha': alpha_true, 'beta': beta_true}[param], c='k')
    plt.ylabel('Frequency')
    plt.title(component)

plt.tight_layout()
```
